```python
import opendatasets as od
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import jovian

%matplotlib inline
```

Creating a copy of dataset to avoid data modification in original dataset

```python
swiggy1_df = swiggy_df.copy()
```

Find the total Null values count in each column.

```python
null_values=swiggy1_df.isna().sum()
null_values
```

Plotting NULL values

```python
null_values = swiggy1_df.isna().sum()
fig = go.Figure(data=[go.Bar(x=null_values.index, y=null_values.values)])
fig.update_layout(title='Null Values Count of Columns in Swiggy1 DataFrame', xaxis_title='Columns', yaxis_title='Count of Null Values')
fig.show()
```

Replace the empty Values or text data with some numeric data : Basic data cleaning

```python
swiggy1_df['rating'] = swiggy1_df['rating'].str.replace('--', '0')
swiggy1_df['cost'] = swiggy1_df['cost'].str.replace('₹', '')
swiggy1_df['rating_count'] = swiggy1_df['rating_count'].str.replace('+','').str.replace('K','000').str.replace('Too Few Ratings', '0').str.replace('ratings', '')
```

Remove all the NULL values

```python

swiggy1_df.dropna(subset=['rating'], inplace=True)
swiggy1_df.dropna(subset=['cost'], inplace=True)
swiggy1_df.dropna(subset=['cuisine'], inplace=True)
```

```python
null_values=swiggy1_df.isna().sum()
null_values
```

changing the data type of the columns for better data processing

```python
swiggy1_df['rating'] = swiggy1_df['rating'].astype(float)
swiggy1_df['rating_count'] = swiggy1_df['rating_count'].astype(int)
swiggy1_df['cost'] = swiggy1_df['cost'].astype(int)
```

```python
swiggy1_df.info()
```

checking whether any null values present or not

```python
swiggy1_df.isna().sum()
```

To view the dataset description

```python
swiggy1_df.describe()
```

Removing outliers in cost column where cost is greater than 10000

```python
swiggy1_df = swiggy1_df.drop(swiggy1_df[swiggy1_df['cost']>10000].index)
```

```python
#Adding a new column of Measures so that we can show it as a label
measures_df = pd.DataFrame(swiggy1_df.describe())
Measures = ['Count', 'Average', 'Std_dev','Minimum','Q1','Q2','Q3','Maximum']
```

```python
measures_df['Central Tendencies'] = Measures
measures_df
```

Plotting Central tendencies for rating column in data frame

```python
#Plotting the central tendencies of rating_count column using a pie chart
fig = px.pie(measures_df,
             values = 'rating_count',
             names = 'Central Tendencies',
             title = 'Central Tendencies of our Ratings column');
fig.show()
```

```python
#Plotting the central tendencies of cost column using a line chart
fig = px.line(measures_df,
             x = 'Central Tendencies',
             y = 'cost',
             title= 'Central Tendencies for Cost');
fig.show()
```

```python
#plotting the cost column
fig = px.box(swiggy1_df,
              y = 'rating',
              title= 'Central Tendencies for rating Column');
fig.show()
```

## visualize Cities that have highest number of rating count given by people

```python
highest_city_rating_count = swiggy1_df.groupby('city')['rating_count'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
highest_city_rating_count.head(10).plot.bar()
plt.xlabel('Cities')
plt.ylabel('Number Of Ratings')
plt.title('Cities that have highest number of rating count given by people')
plt.show();
```

**Banjara Hills** is having the highest rating count and is equal to 12000

## **Top 20 cities as per the number of restaurants listed**

```python
top_rest_df = swiggy1_df.groupby('city')['name'].count().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
top_rest_df.head(20).plot.barh()
plt.xlabel('Number of Restaurants')
plt.ylabel('Cities')
plt.title('Top 20 cities As Per The Number Of Restaurants Listed')
plt.show()
```

Bikaner is the top city having more than 1600 restuarants and second top city comes noida-1 which has more than 1400

## **Relation between Top cities as per the number of restaurants listed and City with highest rating count**

```python
combined_df = pd.merge(top_rest_df, highest_city_rating_count, on = 'city' )
```

```python
fig = px.scatter(combined_df,
                 x = 'rating_count',
                 y = 'name',
                 size = 'rating_count',
                 title = 'Relation Between Number Of Restaurants Listed And Rating Counts',
                 color = combined_df.index)
fig.update_yaxes(title_text='Number Of the Restaurant in City')
fig.update_xaxes(title_text='Number Of the Ratings Given')
fig.show()
```

City Bikaner has the highest rating as well as the highest number of restuarants with 1666 restaurants and 115870 rating_counts
